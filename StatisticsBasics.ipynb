3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
  - Dispersion refers to the extent to which data values are spread out or scattered around a central value (like the mean). It provides insight into the variability, consistency, or diversity within a dataset.
  - Variance: Measures the average squared deviation of each data point from the mean.
  - Standard Deviation: The square root of the variance, expressed in the same units as the data.

4. What is a box plot, and what can it tell you about the distribution of data?
  - A box plot (or box-and-whisker plot) is a graphical representation of data distribution that provides insights into its spread, central tendency, and potential outliers. It is especially useful for visualizing and comparing datasets.

5. Discuss the role of random sampling in making inferences about populations.
  - Role of Random Sampling in Making Inferences:
   - Representation of Population- Random sampling increases the likelihood that the sample reflects the diversity and characteristics of the entire population.
   - Unbiased Estimates: Because random sampling does not favor certain individuals or groups, it reduces systematic errors.

6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
  - Skewness is a statistical measure that describes the asymmetry of a dataset‚Äôs distribution relative to a normal (symmetrical) distribution.
  - Types of Skewness:
    - Positive Skew (Right Skew): The tail on the right side of the distribution is longer.
    - Negative Skew (Left Skew): The tail on the left side of the distribution is longer.
  - How Skewness Affects Data Interpretation:
    - Impact on Central Tendency: In positively skewed data, the mean will be greater than the median due to the influence of extreme values.
    - Effect on Variability: Skewed distributions make measures like standard deviation less informative as outliers disproportionately increase the spread.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
  - The interquartile range (IQR) is a measure of statistical dispersion that describes the spread of the middle 50% of a dataset. It helps identify how data is distributed and is often used to detect outliers.
  - How is the IQR Used to Detect Outliers?
    - Outlier Boundaries: Outliers are data points that fall far outside the range of typical values.
    - Identifying Outliers: Any data point less than the lower bound or greater than the upper bound is considered an outlier.

8. Discuss the conditions under which the binomial distribution is used.
  - The binomial distribution is used to model the probability of observing a specific number of successes in a series of independent trials, where each trial has only two possible outcomes.
  - Conditions for Using the Binomial Distribution:
    - Fixed Number of Trials (n): The experiment must involve a predetermined and finite number of trials.
    - Two Possible Outcomes: Each trial must result in one of two mutually exclusive outcomes (commonly referred to as "success" and "failure").

11. Explain what a random variable is and differentiate between discrete and continuous random variables.
  - A random variable is a numerical representation of the outcomes of a random phenomenon or experiment.
  - 1. Discrete Random Variables:A random variable is discrete if it can take on a finite or countable number of values.
  - 2. Continuous Random Variables: A random variable is continuous if it can take on an infinite number of values within a range.
